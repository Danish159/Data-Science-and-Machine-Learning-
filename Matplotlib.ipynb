```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array([0,6])
y = np.array([0,80])
```

```python
plt.plot(x,y)
plt.show()
```

```python
 # line chart
points = np.array([4,7,2,10,13])
plt.plot(points)
```

```python
a2 = np.array([1,2,3,4,5,6])
plt.plot(a2,linestyle = 'dotted')
```

```python
a2 = np.array([1,2,3,4,5,6])
plt.plot(a2,linestyle = 'dashed')
```

```python

a = np.arange(1,10)
a
```

```python
#bar chart
x = np.array(["x","y","z","w"])
y = np.array([25,30,27,32])
plt.bar(x,y)
plt.show()
```

```python
#scatter plot
x1 =np.array([20,15,11,7,8,10])
y1 = np.array([2,7,15,8,10,12])
plt.scatter(x1,y1)
plt.show()
```

```python
#pie chart
p = np.array([30,2 ,25,20,5,])
plt.pie(p)
```

```python
#histogram
h1 = np.random.normal(170,10,250)
plt.hist(h1)
plt.show()
```
